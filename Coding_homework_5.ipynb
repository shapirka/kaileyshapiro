```python
#2 importing modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3 importing acc_hits and acc_fa data
url = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/acc_fa_hw5.csv'
acc_fa = np.loadtxt(url, delimiter=',')

url2 = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/acc_hits_hw5.csv'
acc_hits = np.loadtxt(url2, delimiter=',')
```

```python
#4 printing the shape of the arrays
print(acc_fa.shape)
print(acc_hits.shape)
```

```python
#6 creating 'time' variable
time = np.linspace(0, 4, 8001)
print(time)
time.shape
```

```python
#7 creating an expression for sampling rate of the data
rate = 8001/4
rate
```

```python
#8 creating subpltos of the first trials of acc_hits and acc_fa against time
fig, ax = plt.subplots(1,2, figsize = (10,5))
ax[0].plot(time, acc_hits[0])
ax[1].plot(time, acc_fa[0])

ax[0].set_xlabel('Time')
ax[1].set_xlabel('Time')
ax[0].set_ylabel('acc_hits -- Voltage')
ax[1].set_ylabel('acc_fa -- Voltage')
```

#### #9 Describing the graphs
* The dominant oscillation for the acc_hits occurs around 3-4s and the acc_fa is around 1 second and again between 2-3 seconds.
* The range for acc_hits is much greater --> -150 to +100 rather the acc_fa only ranges from -30 to +20.
* The activity for acc_fa is has more variability across the smaller range and has multiple peaks/spikes.
* The acc_hits only had strong activity around 2 seconds into the trial but the acc_fa had activity beginning around 1.5 seconds.
* The frequency of acc_fa is much larger than acc_hits.

```python
#10 plotting the ERP for acc_hits and time
acc_hits_mean = np.mean(acc_hits, axis = 0)

plt.plot(time, acc_hits_mean)
plt.plot([2,2], [-20,20], 'k', lw=2)
plt.plot([0,4], [0,0], 'k', lw=1)
plt.xlabel('Time [s]')
plt.ylabel('acc_hits in voltage [\mu V]')
plt.title('ERP for acc_hits')
```

### #11 Interpreting ERP for hits
* There is a ERP that occurs around 3-4 seconds
* The response isn't immediate as there is a delay after the screen touch at 2 seconds.

```python
#12 plotting the ERP for acc_fa and acc_hits on the same graph
fig, ax = plt.subplots(1,2, figsize = (15,5))

acc_fa_mean = np.mean(acc_fa, axis = 0)
ax[0].plot(time, acc_fa_mean)
ax[0].plot([2,2], [-20,20], 'k', lw=2)
ax[0].plot([0,4], [0,0], 'k', lw=1)
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('acc_fa in voltage [\mu V]')
ax[0].set_title('ERP for acc_fa')

ax[1].plot(time, acc_fa_mean, color = 'purple')
ax[1].plot(time, acc_hits_mean, color = 'pink')
ax[1].plot([2,2], [-20,20], 'k', lw=2)
ax[1].plot([0,4], [0,0], 'k', lw=1)
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('voltage [\mu V]')
ax[1].set_title('ERP of acc_fa (purple) and acc_hits (pink)')
```

### #13 Interpreting the overlapping graph in #12
* The hits have a stronger response than the false alarms with more variability.
* The false alarm has less variability with a much smaller amplitude than the hits.
* Before the screen touch at 2 seconds, the data for both acc_hits and acc_fa is pretty similar with basic noise.  The noise remains pretty consistent for the acc_fa but the acc_hits responds to the positive stimulus of a strawberry milkshake (aka an ERP).
