```python
#2 importing packages
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#3 importing data

url = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/v_rest_pvalb.csv'
v_rest_pvalb = np.loadtxt(url)

url2 = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/v_rest_sst.csv'
v_rest_sst = np.loadtxt(url2)
```

```python
#4 finding means of data, performing t-test and plotting histogram

v_rest_sst_mean = np.mean(v_rest_sst)
print(v_rest_sst_mean)
v_rest_pvalb_mean = np.mean(v_rest_pvalb)
print(v_rest_pvalb_mean)

ttest = stats.ttest_ind(v_rest_sst,v_rest_pvalb)
print (ttest)
stats.ttest, p_value = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
print(p_value)

if p_value > 0.05:
  fig, axs = plt.subplots(1,2)
  axs[0].hist(v_rest_sst)
  axs[0].set_xlabel('Resting Potential')
  axs[0].set_ylabel('Frequency')
  axs[1].hist(v_rest_pvalb)
  axs[1].set_xlabel('Resting Potential')
  axs[1].set_ylabel('Frequency')
else:
  print('the p value is <0.05, so significant!')
```

```python
#5 defining vrestHist()

def vrestHist(v_rest_sst,v_rest_pvalb):
  """
  This is a function that will create a histogram depending on the results of the t test for two pieces of data
  """
  v_rest_sst_mean = np.mean(v_rest_sst)
  print(v_rest_sst_mean)
  v_rest_pvalb_mean = np.mean(v_rest_pvalb)
  print(v_rest_pvalb_mean)

  ttest = stats.ttest_ind(v_rest_sst,v_rest_pvalb)
  print (ttest)
  stats.ttest, p_value = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
  print(p_value)

  if p_value > 0.05:
    fig, axs = plt.subplots(1,2)
    axs[0].hist(v_rest_sst)
    axs[0].set_xlabel('Resting Potential')
    axs[0].set_ylabel('Frequency')
    axs[1].hist(v_rest_pvalb)
    axs[1].set_xlabel('Resting Potential')
    axs[1].set_ylabel('Frequency')
  else:
    print('the p value is <0.05, so  significant!')
```

```python
#6 calling vrestHist()

vrestHist(v_rest_sst, v_rest_pvalb)
```

```python
#7 Making scatter plot using a loop
#v_rest = np.concatenate((v_rest_sst,v_rest_pvalb))

for i in range(len(v_rest_sst)):
    sst = v_rest_sst[i]
    pvalb = v_rest_pvalb[i]

    if sst > -60 and pvalb > -60:
        color = 'red'
    elif sst > -60 and pvalb <= -60:
        color = 'green'
    elif sst <= -60 and pvalb <= -60:
        color = 'blue'
    elif sst <= -60 and pvalb >= -60:
        color = 'purple'
    plt.scatter(sst, pvalb, color=color)

plt.xlabel('v_rest_sst')
plt.ylabel('v_rest_pvalb')
plt.show()
```

```python
#8 Creating vrestScatter()

def vrestScatter(v_rest_sst,v_rest_pvalb):
  """
  This function runs a loop through vrest_sst and vrest_pvalb to create a scatter plot based on the following information.
  """
  for i in range(len(v_rest_sst)):
    sst = v_rest_sst[i]
    pvalb = v_rest_pvalb[i]

    if sst > -60 and pvalb > -60:
        color = 'red'
    elif sst > -60 and pvalb <= -60:
        color = 'green'
    elif sst <= -60 and pvalb <= -60:
        color = 'blue'
    elif sst <= -60 and pvalb >= -60:
        color = 'purple'
    plt.scatter(sst, pvalb, color=color)

  plt.xlabel('v_rest_sst')
  plt.ylabel('v_rest_pvalb')
  plt.show()

vrestScatter(v_rest_sst, v_rest_pvalb)
```
