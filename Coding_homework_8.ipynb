```python
#2 Importing Modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
#3 Importing data
url = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/extinction_predictors_hw8.csv'
extinction_predictors = np.loadtxt(url, delimiter = ',')

url2 = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/fear_conditioning_traces_hw8.csv'
fear_conditioning_traces = np.loadtxt(url2, delimiter = ',')

shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter = ',').astype(int)
```

```python
#5 Creating new variables
conditioning_traces = fear_conditioning_traces[:10458,:]
extinction_traces = fear_conditioning_traces[10459:,:]
```

```python
#6 Finding the size of the variables
print(conditioning_traces.shape)
print(extinction_traces.shape)
print(fear_conditioning_traces.shape)
```

```python
#7 Creating time variables
conditioning_rate = 10458/15
print(conditioning_rate)

extinction_rate = 14209/15
print(extinction_rate)
time_conditioning = np.linspace(0, conditioning_rate, 10458)
time_extinction = np.linspace (0, extinction_rate, 14209)
```

```python
#12 Making new variables to model our data
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = conditioning_traces[:,0]

predictors = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
model1 = sm.GLM(first_trace, predictors, family=Poisson())
model1_results = model1.fit() # Fit model to our data
b1 = model1_results.params
print('b1:')
print(b1)
```

```python
#13 Running the model for each ROI in Conditioning_traces
predictors['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r3.statistic

plt.plot(np.sort(r_model_conditioning), np.linspace(0, 1, len(r_model_conditioning), endpoint=False), label = "model 3")
plt.legend()
plt.xlabel('Correlation Between Model Prediction and Observed Trace')
plt.ylabel('Cumulative Proportion of ROIs')
```

```python
#14 Running a model for the Extinction_traces
predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  model_results = model.fit()
  predicted_trace = model_results.predict()
  r = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r.statistic

plt.plot(np.sort(r_model_extinction), np.linspace(0, 1, len(r_model_extinction), endpoint=False), label = "model")
plt.legend()
plt.xlabel('Correlation Between Model Prediction and Observed Trace')
plt.ylabel('Cumulative Proportion of ROIs')
```

```python
#15 Creating a function to correlate the conditioning and extinction variables
r = pearsonr(r_model_conditioning, r_model_extinction)
print(r)
```

```python
#16 Making scatter plot
plt.scatter(r_model_conditioning, r_model_extinction)
plt.xlabel('Model conditioning')
plt.ylabel('Model extinction')
```

### #17 Intrepreting the scatter plot
* From the PearsonR results, the conditioning and extinction values are not significantly related because the p value is >0.05.
* This is also shown in the scatter plot, which does not have a strong correlation among all the plotted points.
* From the scatter plot, it does not seem like a cell that responds to the tone during conditioning would also respond to the tone during extinction.  The conditioning values range from 0.1 to 0.5, rather the extinction values only range from 0.0 to 0.3.  
  * A cell that responds during conditioning with a value of 0.3, would only respond at most with a value of 0.2 during extinction, showing that over time the cells lose their sensitivity to the tone and are less likely to fire.
