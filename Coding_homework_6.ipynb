```python
#2 Importing modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3 Importing urls
url = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/acc_hits_hw6.csv'
acc_hits = np.loadtxt(url, delimiter = ',')

url2 = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/acc_fa_hw6.csv'
acc_fa = np.loadtxt(url2, delimiter = ',')

url3 = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/lc_hits_hw6.csv'
lc_hits = np.loadtxt(url3, delimiter = ',')

url4 = 'https://raw.githubusercontent.com/shapirka/neur_265kaileyshapiro/main/lc_fa_hw6.csv'
lc_fa = np.loadtxt(url4, delimiter = ',')
```

```python
#4 Printing shape of the arrays
print(acc_hits.shape)
print(acc_fa.shape)
print(lc_hits.shape)
print(lc_fa.shape)
```

```python
#6 Creating time variable
time = np.linspace(0, 4, 8001)
time.shape
```

```python
#7 Determing sampling rate
rate = 8001/4
rate
```

```python
#8 Plotting first trial of hits and false alarms in subplots
fig, ax = plt.subplots(1,2, figsize = (10,5))
ax[0].plot(time, acc_hits[0], color = 'r')
ax[0].plot(time, lc_hits[0], color = 'b')

ax[1].plot(time, acc_fa[0], color = 'r')
ax[1].plot(time, lc_fa[0], color = 'b')

ax[0].set_xlabel('Time[s]')
ax[1].set_xlabel('Time[s]')
ax[0].set_ylabel('Voltage')
ax[1].set_ylabel('Voltage')
ax[0].set_title('First Trial Voltage of Acc_hits(red) and Lc_hits (blue)')
ax[1].set_title('First Trial Voltage of Acc_fa(red) and Lc_fa (blue)')
```

### #9 Describing the graphs
* The dominant oscillation for the anterior cingulate cortex hits is roughly 1-1.25 --> about 4-5 tall peaks across 4 seconds.
* The dominant oscillation for the locus coeruleus hits is harder to see but there is maybe 1-2 tall peaks across 4 seconds so it would be about 0.5.
* Comparing the hit and false alarm graphs: there is a much stronger single, isolated peak in the acc_fa graph, rather the acc_hits is more scattered with multiple strong peaks.
* I don't see any phase coherences across these two sets of data as the ACC signals are much stronger than the LC for both hits and false alarms.

```python
#10 Finding the coherence for the hits data
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

T = time[-1]
dt = time[1]-time[0]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

N = acc_hits.shape[1]
f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real)
plt.xlim([0, 50])
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between acc_hits and lc_hits')
```

### #11 Describing the coherence graph
* There is the strongest level of coherence around 5-10 Hz as this is when the graph is closet to 1, indicating it has a strong level of coherence.

```python
#12 Creating coherence graph for false alarms
#hits:
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

T = time[-1]
dt = time[1]-time[0]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

N = acc_hits.shape[1]
f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real, color = 'r')
plt.xlim([0, 50])
plt.ylim([0, 1])


#false alarms:
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])

T = time[-1]
dt = time[1]-time[0]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

N = acc_fa.shape[1]
f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real, color = 'b')
plt.xlim([0, 50])
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between hits (red) and false alarms (blue)')
```

### #13 Interpreting the graph
* The coherence is slightly higher for the hits compared to the false alarms.
* Both hits and false alarms have the strongest coherence at ~1 Hz and also a little before 10 Hz (around 6-8ish).
*The stronger coherence indicates more communication between the anterior cingulate cortex and the locus coeruleus so this means there is stronger synaptic connections with the hits since those lines are always slightly above the blue/false alarm lines.
